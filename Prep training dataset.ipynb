```python
import boto3
import json 
import pandas as pd


bedrock_runtime =boto3.client(
    service_name="bedrock-runtime",
    region_name ="us-east-1"
)
```

```python
prompt = 'Create a unique medical note and make it greater 800 words long, create random names, conditions, procedures. etc. '
```

```python
kwargs = {
  "modelId": "ai21.j2-mid-v1",
  "contentType": "application/json",
  "accept": "*/*",
  "body": "{\"prompt\":\""+ prompt + "\",\"maxTokens\":3000,\"temperature\":0.1,\"topP\":1,\"stopSequences\":[],\"countPenalty\":{\"scale\":0},\"presencePenalty\":{\"scale\":0},\"frequencyPenalty\":{\"scale\":0}}"
}

#print(kwargs)
```

```python
sample_note = []
for i in range(9):
    response = bedrock_runtime.invoke_model(**kwargs)
    response_body = json.loads(response.get('body').read())
    completion = response_body.get('completions')[0].get('data').get('text')
    sample_note.append(completion)
    
```

```python
for s in sample_note:
    print (s)
    print ('------------------------')
```

```python
df = pd.DataFrame({'Notes': sample_note})
pd.set_option('display.max_colwidth', None)

# Display the DataFrame
df
```

```python
def create_summary(data): 
    povider_note = data 
    trainning_sam_resp_prompt = f"Write a concise summary of the main points and key insights from this provider note: {povider_note}".encode('unicode_escape').decode('utf-8') 
    trainning_sam_resp_kwargs = {
        "modelId": "ai21.j2-mid-v1",
        "contentType": "application/json",
        "accept": "*/*",
        "body": "{\"prompt\":\""+ trainning_sam_resp_prompt + "\",\"maxTokens\":3000,\"temperature\":0.1,\"topP\":1,\"stopSequences\":[],\"countPenalty\":{\"scale\":0},\"presencePenalty\":{\"scale\":0},\"frequencyPenalty\":{\"scale\":0}}"
    }
    
    trainning_sam_resp_response = bedrock_runtime.invoke_model(**trainning_sam_resp_kwargs)
    trainning_sam_resp_response_body = json.loads(trainning_sam_resp_response.get('body').read())
    trainning_sam_resp_completion = trainning_sam_resp_response_body.get('completions')[0].get('data').get('text')
    return trainning_sam_resp_completion
```

```python
df['output_text'] = df["Notes"].apply(create_summary)
```

### Dataset for Tuning

- Divide the data into a training and evaluation. By default, 80/20 split is used.
- This (80/20 split) allows for more data to be used for tuning. The evaluation split is used as unseen data during tuning to evaluate performance.
- The `random_state` parameter is used to ensure random sampling for a fair comparison. 

```python
from sklearn.model_selection import train_test_split
```

```python
train, evaluation = train_test_split(
    df,
    ### test_size=0.2 means 20% for evaluation
    ### which then makes train set to be of 80%
    test_size=0.2,
    random_state=42
)
```
